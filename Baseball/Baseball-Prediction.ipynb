```python
sns.jointplot(data.R, data.W)
plt.xlabel("Runs")
plt.ylabel("No. of Wins")
```

```python
sns.jointplot(data.AB, data.W)
plt.xlabel("At bats")
plt.ylabel("No. of Wins")
```

```python
sns.jointplot(data.HR, data.W)
plt.xlabel("Home Run")
plt.ylabel("No. of Wins")
```

```python
sns.jointplot(data["2B"], data.W)
plt.xlabel("Double PLay")
plt.ylabel("No. of Wins")
```

```python
sns.jointplot(data.E, data.W)
plt.xlabel("Error")
plt.ylabel("No. of Wins")
```

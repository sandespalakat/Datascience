```python
#Outlier removal
```

```python
from scipy.stats import zscore
```

```python
zs = np.abs(zscore(data))
```

```python
zs
```

```python
print(np.where(zs > 3))
```

```python
baseball = data[(zs < 3).all(axis=1)]
```

```python
baseball.skew()
```

```python
# Skewness still exists in columns HIts,CG,saves and Error
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
columns = ["H", "CG", "SV", "E"]
for column in columns:
    pt = PowerTransformer()
    pt.fit(np.array(baseball[column]).reshape(-1,1))
    baseball[column] = pt.transform(np.array(baseball[column]).reshape(-1,1))
```

```python
baseball.skew()
```

```python
#skewness is removed
```

```python
x = baseball.iloc[:,1:]
```

```python
y = baseball.iloc[:,1]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
#finding best random state
max_acc = 0
max_rs = 0
for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .22, random_state = i)
    lr = LinearRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    r2 = r2_score(y_test, pred)
    if r2 > max_acc:
        max_acc = r2
        max_rs = i
print("Best random state=",max_rs)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .22, random_state = i)
```

```python
models = {"SVR":SVR(),"KNN":KNeighborsRegressor(), "RandomForest":RandomForestRegressor(),"LinearRegression":LinearRegression(), "Ridge":Ridge(), "dtr":DecisionTreeRegressor() }
acc = {}
mod_list = []
for i in models:
    mod = i
    mod = models[i]
    #mod = DecisionTreeRegressor()
    mod.fit(x_train, y_train)
    pred = mod.predict(x_test)
    r2_sc = r2_score(y_test,pred)
    acc[i] = r2_sc
    mod_list.append(mod)
print(acc)
```

```python
#applying cross validation
from sklearn.model_selection import cross_val_score
```

```python
cv_list = {}
for i in mod_list:
    cv = cross_val_score(i, x, y, cv=7)
    cv_list[str(i)] = cv.mean()
print(cv_list)
```

```python
#Linear regression is the best algorithm here
```
